## Oscilador Amortecido

Um corpo tem seu movimento descrito pela função posição :

$$x(t) = A \exp\left(\frac{-b\,t}{2\, m}\right) \cos(\omega\,t)$$

Considere $A = 3,0\,m$, $b = 2,0\,kg/s$, $m = 25\,kg$ e $\omega = \pi\,rad/s$ :

### 1) Gráfico de $x(t)$

```python
from sympy import *
```

```python
A, b, omega, m, t = symbols('A, b, omega, m, t')
```

```python
def posicaox(A, b, m, omega, t):
    return A*exp(-b*t/(2*m))*cos(omega*t)
```

```python
posicaox(A, b, m, omega, t)
```

```python
posicaox(3, 2, 25, pi, t)
```

```python
posicaox(3, 2, 25, pi, 5.0)
```

```python
posicaox(3, 2, 25, pi, 70.0)
```

```python
plot(posicaox(3, 2, 25, pi, t), (t, 0, 30));
```

A gráfico da posição mostra que a amplitude inicial é $3,0\,m$, porém vai caindo com o tempo no oscilador amortecido.

```python
plot(posicaox(3, 2, 25, pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='x(m)', 
  title=u'Posição do Oscilador Amortecido');
```

Com $b$ maior, a oscilação fica mais amortecida :

```python
plot(posicaox(3, 5, 25, pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='x(m)', 
  title=u'Posição do Oscilador Amortecido');
```

Aumentando $m$ (massa), diminui o amortecimento da oscilação :

```python
plot(posicaox(3, 2, 100, pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='x(m)', 
  title=u'Posição do Oscilador Amortecido');
```

Aumentando $\omega$ (frequência angular), aumenta o número de oscilações por período de tempo :

```python
plot(posicaox(3, 2, 100, 2*pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='x(m)', 
  title=u'Posição do Oscilador Amortecido');
```

### 2) Velocidade $v(t)$

```python
diff(posicaox(A, b, m, omega, t), t)
```

```python
diff(posicaox(A, b, m, omega, t), t, 15)
```

```python
def velocidade(A, b, m, omega, t):
    return diff(posicaox(A,b,m,omega,t), t)
```

```python
velocidade(A, b, m, omega, t)
```

```python
plot(velocidade(3, 2, 25, pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='v(m/s)', 
  title=u'Velocidade do Oscilador Amortecido');
```

```python
velocidade(3, 2, 25, pi, t)
```

### 3) Aceleração $a(t)$

```python
diff(posicaox(A, b, m, omega, t), t, 2)
```

```python
def aceleracao(A, b, m, omega, t):
    return diff(posicaox(A,b,m,omega,t), t, 2)
```

```python
aceleracao(A, b, m, omega, t)
```

```python
plot(aceleracao(3, 2, 25, pi, t), (t, 0, 33), 
  xlabel='t(s)', ylabel='$a(m/s^2)$', 
  title=u'Aceleração do Oscilador Amortecido');
```

```python
aceleracao(3, 2, 25, pi, t)
```

### 4a Questão : cálculo de $K=m v^2/2$ e seu gráfico

A energia cinética $K=m v^2/2$ é obtida a partir da derivada primeira da função posição $x(t)$ em relação a $t$ :

```python
m*diff(posicaox(A,b,m,omega,t),t)**2/2
```

```python
expand(m*diff(posicaox(A,b,m,omega,t),t)**2/2)
```

```python
simplify(m*diff(posicaox(A,b,m,omega,t),t)**2/2)
```

```python
plot(30*diff(posicaox(5, 2, 30, pi, t),t)**2/2,(t,0,30),
     xlabel='t',ylabel='K',
     title=u'Energia cinética do oscilador amortecido');
```

A energia cinética é sempre positiva e começa nula, oscila e sua amplitude diminui com o tempo.

### Vetores em Python com SymPy para vetores numéricos e simbólicos

Função "Matrix" define um vetor (que é um caso específico de uma matriz, aqui um vetor coluna 3x1) :

"dot" é um método do vetor para calcular o produto escalar :

```python
vetor4.dot(vetor4)
```

"cross" é um método do vetor para calcular o produto vetorial :

```python
vetor4.cross(vetor4)
```

Vetores com elementos simbólicos :

```python
t, R, omega = symbols("t, R, omega")
```

Vetor posição de um movimento circular uniforme em 2D :

```python
r = Matrix(2,1,[R*cos(omega*t),R*sin(omega*t)])
```

```python
r
```

```python
r.dot(r)
```

```python
simplify(r.dot(r))
```

Vetor velocidade é a derivada do vetor posição em relação ao tempo t, onde "diff" é um método do vetor :

```python
v = r.diff(t); v
```

```python
simplify(v.dot(v))
```

Vetor posição r é ortogonal ao vetor velocidade v :

```python
r.dot(v)
```

Vetor aceleração é a derivada do vetor velocidade em relação ao tempo t :

```python
a = v.diff(t); a
```

```python
simplify(a.dot(a))
```

Ou seja, $a = v^2/R$ :

```python
simplify(v.dot(v))/R
```

Vetor velocidade v é ortogonal ao vetor aceleração a :

```python
v.dot(a)
```
